####**4.1. Aplicación de Estadísticos Descriptivos Básicos.**

```python
# Descriptivos Básicos de los Datos:

data_frame.describe()
```

```python
# Descriptivos Básicos por Tipo de Variable:

print('Mínimo:',data_frame['company'].min())
print('Máximo:',data_frame['company'].max())
print('Promedio:',data_frame['company'].mean())
print('STD:',data_frame['company'].std())
print(data_frame.company.quantile([.25, .5, .75]))
print('*******')
print('Mínimo:',min(data_frame['company']))
print('Máximo:',max(data_frame['company']))
```

```python
# Considerando Solamente las Variables Object:

data_frame.describe(include=[object])
```

```python
fig, axes = plt.subplots(nrows=13, ncols=4, figsize=(20, 35))
axes = axes.flat
columnas_numeric = data_frame.select_dtypes(include=['float64', 'int']).columns

for i, colum in enumerate(columnas_numeric):
    sns.histplot(
        data    = data_frame,
        x       = colum,
        stat    = "count",
        kde     = True,
        color   = (list(plt.rcParams['axes.prop_cycle'])*9)[i]["color"],
        line_kws= {'linewidth': 2},
        alpha   = 0.3,
        ax      = axes[i],
           )
    axes[i].set_title(colum, fontsize = 9, fontweight = "bold")
    axes[i].tick_params(labelsize = 8)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("")


fig.tight_layout()
plt.subplots_adjust(top = 0.95)
fig.suptitle('Distribución de Variables Numéricas', fontsize = 12, fontweight = "bold");
```

####**4.2. Análisis de Tendencias y Comportamientos de los Datos.**

```python
# Transformación de Variable hotel en Numérica:

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
data_frame["hotel"] = lb_make.fit_transform(data_frame["hotel"])
data_frame[["hotel"]].head(10)
```

```python
# Transformación de Variables Categóricas en Numéricas:

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
data_frame["company"] = lb_make.fit_transform(data_frame["company"])
data_frame[["company"]].head(10)
```

```python
data_frame[["company"]].describe()
```

Veremos la descripcion de la variable hotel, la cual nos describe el conteo total de oteles y cual es el top entre ellos.

```python
data_frame[["hotel"]].describe()
```

```python
# Conjunto de Datos con la Variable hotel Transformada:

data_frame
```

```python
result = data_frame.groupby('company')['hotel'].max()
result
```

```python
# Gráficas de Distribución de Variables Categóricas:

data_frame.groupby('company').size().plot(kind='bar', figsize = (10, 5))
plt.title('company')
plt.show()
```

¿Cuantos vuelos son cancelados por CADA UNA de las diferentes compañias?

```python
result = data_frame.groupby('company')['is_canceled'].max()
result
```

¿Cuantos vuelos arriban cada mes?

```python
result = data_frame.groupby('arrival_date_month')['company'].max()
result
```

¿Cuantos vuelos arriban cada año?

```python
result = data_frame.groupby('arrival_date_year')['company'].count()
result
```
